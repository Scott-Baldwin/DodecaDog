PLOTS

```python
import numpy as np
from snippets import QuickScatter

n = 50
x = np.random.rand(n)
y = np.random.rand(n)
z = np.random.rand(n)

fig, ax = QuickScatter(x,y,z)
```

```python
fig, ax = QuickScatter(x,y,z,
        cm='cool', 
        vmin=-.5, 
        vmax=1.5, 
        ticks=[0, 1],
        xlabel='Random X',
        ylabel='Random Y',
        units='Units',
        pad=0.2)
```
